#### Applying XGBoost algorithm on the Boston housing dataset to predict the median value. Do the usual train/test split and then compare the result with the result of applying linear regression. You can use a suitable metric like RMSE.



```python
import pandas as pd
import numpy as np
import seaborn as sns
from xgboost import XGBRegressor
from xgboost import plot_tree
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
#conda install -c conda-forge xgboost
```

```python
df = pd.read_excel('/Users/wadhwa/Desktop/dsem/Assignment-6/boston housing.xlsx')
df
```

```python
df.isnull()
```

```python
# Heatmap to visualize missing values
sns.set(style="white")
plt.figure(figsize=(8,6))
sns.heatmap(df.isnull(), cmap="Reds", cbar=True)
plt.title("Missing Values Heatmap", fontsize=14, fontweight='bold')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap="BuPu")
plt.title("Correlation Heatmap", fontsize=14, fontweight='bold')
plt.show()
```

```python
X = df.drop('MV', axis=1)
Y = df['MV']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=7)
```

```python
xgb_reg_model = XGBRegressor(learning_rate=0.1, max_depth = 5, objective ='reg:linear', alpha = 4)
xgb_reg_model.fit(X_train, y_train)
```

```python
#conda install graphviz
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x=xgb_reg_model.feature_importances_, y=X.columns, palette='Blues_r')
plt.title("Feature Importance Plot", fontsize=16, fontweight='bold')
plt.xlabel("Importance", fontsize=14)
plt.xticks(fontsize=12)
plt.ylabel("Features", fontsize=14)
plt.yticks(fontsize=12)
plt.show()
```

```python
y_pred_xgb = xgb_reg_model.predict(X_test)
```

```python
rmse_xgb = np.sqrt(mean_squared_error(y_test, y_pred_xgb))
print("RMSE for XGBoost: ", rmse_xgb)
```

```python
lin_reg_model = LinearRegression()
lin_reg_model.fit(X_train, y_train)
```

```python
y_pred_lin = lin_reg_model.predict(X_test)
```

```python
plt.figure(figsize=(8, 6))
sns.residplot(y_test, y_pred_xgb, color="orange")
plt.title("XGBoost Residuals Plot", fontsize=14, fontweight='bold')
plt.xlabel("True Values", fontsize=12)
plt.ylabel("Residuals", fontsize=12)
plt.show()

plt.figure(figsize=(8, 6))
sns.residplot(y_test, y_pred_lin, color="brown")
plt.title("Linear Regression Residuals Plot", fontsize=14, fontweight='bold')
plt.xlabel("True Values", fontsize=12)
plt.ylabel("Residuals", fontsize=12)
plt.show()
```

```python
rmse_lin = np.sqrt(mean_squared_error(y_test, y_pred_lin))
print("RMSE for Linear Regression: ", rmse_lin)
```

```python
plt.figure(figsize=(8,6))
plt.bar(['XGBoost', 'Linear Regression'], [rmse_xgb, rmse_lin], color=['#FFA500', '#8B4513'])
plt.title("Comparison of RMSE", fontsize=16, fontweight='bold')
plt.xlabel("Model", fontsize=14)
plt.xticks(fontsize=12)
plt.ylabel("RMSE", fontsize=14)
plt.yticks(fontsize=12)
plt.ylim(0, max(rmse_xgb, rmse_lin)*1.1)
for i, value in enumerate([rmse_xgb, rmse_lin]):
    plt.text(i, value+0.05, str(round(value, 2)), ha='center', fontsize=12, fontweight='bold')
plt.show()
```

```python
# Plot actual vs predicted plot for XGBoost model
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred_xgb, color='#007acc', alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel("Actual Values", fontsize=12)
plt.ylabel("Predicted Values", fontsize=12)
plt.title("Actual vs Predicted - XGBoost", fontsize=14, fontweight='bold')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()

# Plot actual vs predicted plot for Linear Regression model
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred_lin, color='#ff7f0e', alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel("Actual Values", fontsize=12)
plt.ylabel("Predicted Values", fontsize=12)
plt.title("Actual vs Predicted - Linear Regression", fontsize=14, fontweight='bold')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```
